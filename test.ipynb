```python
climbing_stairs_data = data[data["problemTitle"] == "Climbing Stairs"]
third_max_data = data[data["problemTitle"] == "Third Maximum Number"]
two_sum_data = data[data["problemTitle"] == "Two Sum"]

# Save into separate files
climbing_stairs_data.to_csv("climbing_stairs.csv", index=False)
third_max_data.to_csv("third_maximum_number.csv", index=False)
two_sum_data.to_csv("two_sum.csv", index=False)
```

```python
climbing_stairs_data_c=climbing_stairs_data[climbing_stairs_data["language"] == "c"]
climbing_stairs_data_c.to_csv("climbing_stairs_c.csv", index=False)
```

```python
climbing_stairs_data_cpp=climbing_stairs_data[climbing_stairs_data["language"] == "cpp"]
climbing_stairs_data_cpp.to_csv("climbing_stairs_cpp.csv", index=False)
```

```python
climbing_stairs_data_java=climbing_stairs_data[climbing_stairs_data["language"] == "java"]
climbing_stairs_data_java.to_csv("climbing_stairs_java.csv", index=False)
```

```python
climbing_stairs_data_python=climbing_stairs_data[climbing_stairs_data["language"] == "python"]
climbing_stairs_data_python.to_csv("climbing_stairs_python.csv", index=False)
```
